The attached document (your puzzle input) contains a sequence of rotations, one per line, which tell you how to open the safe. A rotation starts with an L or R which indicates whether the rotation should be to the left (toward lower numbers) or to the right (toward higher numbers). Because the dial is a circle, turning the dial left from 0 one click makes it point at 99. Similarly, turning the dial right from 99 one click makes it point at 0.

# Write function to calculate rotations past zero

```python
# Separate letter from numbers for caluculation- within fn or before?
# Create zero counter, if arithmetic hits zero or 100, + 1
# R = addition, L = subtraction
# Create ticker starting at 50 rest in while loop
    # if it goes negative, add 100 and put result into ticker (also add to zero counter if exactly zero)
    # if it goes over 99, subtract 100 and put result into ticker (also add to zero counter if exactly 100)

def zero_ticker(input):

    # Create counter for each time ticker hits zero
    zero_counter = 0
    # Create rotation index to iterate through list of rotations
    rotation_index = 0
    # Create ticker to sum rotations (start at 50 per instructions)
    ticker = 50
    # Iterate through list of rotations
    while rotation_index < len(input):
        rotation = input[rotation_index]
        # Split the letter from the number
        letter = rotation[:1]
        number = rotation[1:]
        if letter == 'L':
            ticker -= int(number)
            while ticker < 0:
                ticker = ticker + 100
            if ticker == 0:
                zero_counter += 1
        elif letter == 'R':
            ticker += int(number)
            while ticker > 99:
                ticker = ticker - 100
            if ticker == 0:
                zero_counter += 1
                ticker = ticker - 100
        rotation_index += 1

    return zero_counter
```

```python
zero_ticker(rotations)
```

### Description Part 2

You're sure that's the right password, but the door won't open. You knock, but nobody answers. You build a snowman while you think.

As you're rolling the snowballs for your snowman, you find another security document that must have fallen into the snow:

"Due to newer security protocols, please use password method 0x434C49434B until further notice."

You remember from the training seminar that "method 0x434C49434B" means you're actually supposed to count the number of times any click causes the dial to point at 0, regardless of whether it happens during a rotation or at the end of one.

Following the same rotations as in the above example, the dial points at zero a few extra times during its rotations:

The dial starts by pointing at 50.
The dial is rotated L68 to point at 82; during this rotation, it points at 0 once.
The dial is rotated L30 to point at 52.
The dial is rotated R48 to point at 0.
The dial is rotated L5 to point at 95.
The dial is rotated R60 to point at 55; during this rotation, it points at 0 once.
The dial is rotated L55 to point at 0.
The dial is rotated L1 to point at 99.
The dial is rotated L99 to point at 0.
The dial is rotated R14 to point at 14.
The dial is rotated L82 to point at 32; during this rotation, it points at 0 once.
In this example, the dial points at 0 three times at the end of a rotation, plus three more times during a rotation. So, in this example, the new password would be 6.

Be careful: if the dial were pointing at 50, a single rotation like R1000 would cause the dial to point at 0 ten times before returning back to 50!

Using password method 0x434C49434B, what is the password to open the door?

```python
# Same thing but make it so + 1 to zero_counter each time it passes zero
# Don't foget to +1 for each move past zero or 100 during same rotation

def zero_ticker_plus(input):

    # Create counter for each time ticker hits zero
    zero_counter = 0
    # Create rotation index to iterate through list of rotations
    rotation_index = 0
    # Create ticker to sum rotations (start at 50 per instructions)
    ticker = 50
    # Iterate through list of rotations
    while rotation_index < len(input):
        rotation = input[rotation_index]
        # Split the letter from the number
        letter = rotation[:1]
        number = rotation[1:]
        # If dial is moving to left
        if letter == 'L':
            # Don't count the reset in the ticker function
            if ticker == 0:
                zero_counter -= 1
            # Subtract ticker - integer to move dial to left
            ticker -= int(number)
            # Add to zero count each time dial moves past zero
            while ticker < 0:
                ticker = ticker + 100
                zero_counter += 1
            # If ticker is zero, add to ticker counter
            if ticker == 0:
                zero_counter += 1
        # If dial is moving to right:
        elif letter == 'R':
            # Add ticker + integer to move dial to right
            ticker += int(number)
            # Add to zero count each time dial moves past zero
            while ticker > 99:
                ticker = ticker - 100
                zero_counter += 1
                # Don't count the reset in the ticker function
                if ticker == 0:
                    zero_counter -= 1
            # If ticker is zero, add to ticker counter
            if ticker == 0:
                zero_counter += 1
        rotation_index += 1

    return zero_counter
```
